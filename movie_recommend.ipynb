# Preprocessing

```python
import ast     # converts string of lists into a list
```

```python
def convert_1(obj):        # fetch genres and keywords
    g = []
    for j in ast.literal_eval(obj):
        g.append(j['name'])
    return g    
```

```python
data['genres'] = data['genres'].apply(convert_1)
```

```python
data['keywords'] = data['keywords'].apply(convert_1)
```

```python
data
```

```python
def convert_2(obj):              # fetch 3 cast/actors' names
    n = 0
    c = []
    for i in ast.literal_eval(obj):
        if n <= 3:
            c.append(i['name'])
            n = n+1
    return c        
```

```python
data['cast'] = data['cast'].apply(convert_2)
```

```python
def convert_3(text):            # fetch director's name
    d = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            d.append(i['name'])
    return d 
```

```python
data['crew'] = data['crew'].apply(convert_3)
```

```python
#data.iloc[119].crew
```

```python
data.head(3)
```

```python
# convert string into list for overview column

data['overview'] = data['overview'].apply(lambda x:x.split())  
```

```python
data
```

```python
def join(obj):               # join spaces between the names
    p=[]
    for i in obj:
        p.append(i.replace(" ",""))
    return p    
```

```python
data['cast'] = data['cast'].apply(join)
data['crew'] = data['crew'].apply(join)
data['keywords'] = data['keywords'].apply(join)
data['genres'] = data['genres'].apply(join)
```

```python
data.head()
```

# merge into a single column

```python
data['info'] = data['overview'] + data['keywords'] + data['genres']+ data['cast'] + data['crew']
```

```python
data2 = data[['id', 'original_title', 'info']]
```

```python
data2.head()
```

```python
# convert list into string
data2['info'] = data2['info'].apply(lambda x:' '.join(x))
```

```python
# convert into lowercase
data2['info'] = data2['info'].apply(lambda x:x.lower())
```

```python
data2['info'][119]
```

# Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
```

```python
X = cv.fit_transform(data2['info']).toarray()
```

```python
X
```

```python
cv.get_feature_names()
```

# cosine_similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
dist = cosine_similarity(X)
```

# movie_recommend

```python
def movie_recommend(title):
    index = data2[data2['original_title'] == title].index[0]
    movies_list = sorted(list(enumerate(dist[index])), reverse= True, key = lambda x: x[1])[1:6]
    
    for i in movies_list:
        print(data2.iloc[i[0]]['original_title'])
```

```python
movie_recommend('Spider-Man 3')
```

```python
movie_recommend('Harry Potter and the Half-Blood Prince')
```
